```python
from tensorflow.keras.datasets import imdb

from tensorflow.keras.preprocessing import sequence

from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import Embedding ,SimpleRNN , Dense
```

```python
from tensorflow.keras.preprocessing import sequence
```

```python
max_lenth=500
x_train=sequence.pad_sequences(x_train,maxlen=max_lenth)
```

```python
x_train.shape
```

```python
# creating model
```

```python
dim=10
```

```python
voc_size=10000
max_len=20
```

```python
model=Sequential()
model.add(Embedding(max_features,128,input_length=max_len))
model.add(SimpleRNN(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
#create an instance of early_stopping_callback
```

```python
from tensorflow.keras.callbacks import EarlyStopping

early_stopping=EarlyStopping(monitor='val_loss',patience=10,restore_best_weights=True)
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,batch_size=32,validation_split=0.2,callbacks=early_stopping,epochs=10)
```

```python
model.save("simple_rnn_imdb.h5")
```
