```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
## Define the vocabulary size
voc_size=10000
```

```python
sent
```

```python
### One Hot Representation
one_hot_rep=[one_hot(word,voc_size) for word in sent]
one_hot_rep
```

```python
## word Embedding Representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sen_length=8
embedded_doc=pad_sequences(one_hot_rep,padding='pre',maxlen=sen_length)
```

```python
embedded_doc
```

```python
dim=10
```

```python
model=Sequential()
model.add(Embedding(voc_size,dim,input_length=sen_length))
```

```python
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
